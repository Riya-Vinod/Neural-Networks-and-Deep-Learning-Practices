```python
import numpy as np
# XOR dataset
X = np.array([[0, 0],
 [0, 1],
 [1, 0],
 [1, 1]])
y = np.array([-1, 1, 1, -1])
# Initialize weights for two ADALINE units
weights_1 = np.random.rand(2)
weights_2 = np.random.rand(2)
bias_1 = np.random.rand()
bias_2 = np.random.rand()
learning_rate = 0.01
epochs = 30
# Training MADALINE
for epoch in range(epochs):
    for i in range(len(X)):
        # Net inputs
        net1 = np.dot(X[i], weights_1) + bias_1
        net2 = np.dot(X[i], weights_2) + bias_2
        # Outputs of hidden units
        out1 = 1 if net1 >= 0 else -1
        out2 = 1 if net2 >= 0 else -1
        # Majority vote
        final_output = 1 if (out1 + out2) >= 0 else -1
        # Error correction
        if final_output != y[i]:
            error = y[i] - final_output
            weights_1 += learning_rate * error * X[i]
            weights_2 += learning_rate * error * X[i]
    print(f"Epoch {epoch+1} completed")
# Testing MADALINE
print("\nTesting MADALINE:")
for i in range(len(X)):
    net1 = np.dot(X[i], weights_1) + bias_1
    net2 = np.dot(X[i], weights_2) + bias_2
    out1 = 1 if net1 >= 0 else -1
    out2 = 1 if net2 >= 0 else -1
    final_output = 1 if (out1 + out2) >= 0 else -1
    print(f"Input: {X[i]} Output: {final_output}")
```
